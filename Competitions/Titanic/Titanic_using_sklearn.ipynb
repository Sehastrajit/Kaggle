```python
X_train = dtrain.iloc[:,1:].values
X_train
```

```python
y_train = dtrain.iloc[:,0].values
y_train
```

```python
from sklearn.preprocessing import OrdinalEncoder
import numpy as np

enc = OrdinalEncoder()
enc.fit(X_train)
z=enc.transform(X_train)
z
```

```python
zt=pd.DataFrame(z)
ztrain2 = (zt.fillna(0)).values
ztrain2
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt


clf = LogisticRegression()
clf.fit(ztrain2, y_train)
```
